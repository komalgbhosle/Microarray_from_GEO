```python
def getGEO(GSEID):
    try:
        gse = GEOparse.get_GEO(geo = GSEID, destdir = "/data/users-workspace/ruchika.sharma/discoveric/rna_seq/rna_seq_final/klebseilla_pneumonia/analysis/gene_expression_results", silent=True)
        return gse
    except Exception as e:
        print(GSEID)
        print(e)
```
