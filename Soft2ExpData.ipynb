```python
import pandas as pd
import numpy as np
import GEOparse, re, copy
```

```python
def getGEO(GSEID):
    try:
        gse = GEOparse.get_GEO(geo = GSEID, destdir = "/data/users-workspace/ruchika.sharma/Alzheimers/Microarray", silent=True)
        return gse
    except Exception as e:
        print(GSEID)
        print(e)
```

```python
GSEID = "GSE30945"
gse = getGEO(GSEID)
```

```python
#For SampInfo
#possibilities 1.source_name_ch1 2.title 3.characteristics_ch1.0.status, characteristics_ch1.3.agent
df_sampInfo = copy.deepcopy(gse.phenotype_data)
df_sampInfo = df_sampInfo.reset_index(drop=True).rename(columns={'geo_accession':'SampleID','series_id':'ExperimentID'})
```

```python
df_sampInfo.head()
```

```python
df_sampleInfo = df_sampInfo[['ExperimentID', 'SampleID', 'source_name_ch1']]
df_sampleInfo.head()
```

```python
df_sampleInfo['source_name_ch1'] = df_sampleInfo['source_name_ch1'].replace('Healthy Control .+','Healthy',regex=True)
df_sampleInfo['source_name_ch1'] = df_sampleInfo['source_name_ch1'].replace('Alzheimer`s Disease .+','Knockout',regex=True)

df_sampleInfo
```

```python
df_sampleInfo.to_csv('/data/users-workspace/ruchika.sharma/Alzheimers/Microarray/GSE30945_Sample_info.csv', index = False)
```

```python
#For expression data
gpl_name = ([key for key in gse.gpls.keys()])[0]
pivoted_samples = gse.pivot_samples('VALUE')
samples = gse.pivot_samples("VALUE")
```

```python
gse.pivot_samples('VALUE')
```

```python
gse.gpls[gpl_name].table.head(50)
```

```python
gse_annotated = pivoted_samples.reset_index().merge(gse.gpls[gpl_name].table[["ID","Gene_Symbol"]], left_on='ID_REF', right_on="ID").set_index('ID_REF')
# gse_annotated = pivoted_samples.reset_index().merge(df[["ID", "Gene Symbol"]], left_on='ID_REF', right_on="ID").set_index('ID_REF')
```

```python
gse_annotated.reset_index(level=0, inplace=True)
gse_annotated.dropna()
```

```python
del gse_annotated["ID_REF"]
del gse_annotated["ID"]
gse_annotated = gse_annotated.dropna(subset=["Gene_Symbol"])#Replace
gse_annotated = gse_annotated.rename(columns={"Gene_Symbol": "Gene"})
gse_annotated = gse_annotated[~gse_annotated.Gene.str.contains("///")]
#gse_annotated['Gene'] = gse_annotated['Gene'].str.split('//').str[1]
gse_annotated = gse_annotated.groupby("Gene").median()
gse_annotated = gse_annotated.apply(np.log2, axis=0)
gse_annotated.reset_index(level=0, inplace=True)
#gse_annotated['Gene'] = gse_annotated['Gene'].str.split('//').str[1]
```

```python
gse_annotated
```

```python
gse_annotated = gse_annotated.iloc[1:]
# #gse_annotated = gse_annotated[gse_annotated['Gene'] != '---']
# #gse_annotated = gse_annotated.iloc[1:]
gse_annotated.reset_index(drop=True, inplace=True)
gse_annotated
```
